```python
# Create a list of years
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]  

# Create a figure and a gridspec layout with multiple rows and columns
num_rows = (len(years) + 2) // 3
fig = plt.figure(figsize=(18, 8 * num_rows))  
gs = gridspec.GridSpec(num_rows, 3, width_ratios=[1, 1, 1])

for i, year in enumerate(years):
    # Filter the data for the current year
    migrant_df_year = migrant_df[migrant_df['Incident year'] == year]

    # Group and sum the data by 'Reported Month'
    monthly_sum = migrant_df_year.groupby('Reported Month')['Total Number of Dead and Missing'].sum()

    # Convert the month names to datetime format to ensure correct ordering
    month_order = pd.to_datetime(monthly_sum.index, format='%B').sort_values()

    # Reorder the monthly_sum based on the sorted month_order
    monthly_sum_sorted = monthly_sum.loc[month_order.strftime('%B')]

    # Calculate the row and column indices
    row_idx = i // 3
    col_idx = i % 3

    # Create a subplot within the specified gridspec position
    ax = plt.subplot(gs[row_idx, col_idx])
    ax.plot(monthly_sum_sorted.index, monthly_sum_sorted.values, marker='o', color='green')
    ax.set_xlabel('Reported Month')
    ax.set_ylabel('Total Number of Dead and Missing')
    ax.set_title(f'Year {year} - Aggregated by Month', fontweight='bold', fontsize=14, color='blue')
    ax.set_xticks(monthly_sum_sorted.index)
    ax.set_xticklabels(monthly_sum_sorted.index, rotation=45, ha='right')  # Adjust rotation and alignment
    
    # Annotate data values directly on the line plot
    for j, (month, value) in enumerate(monthly_sum_sorted.items()):
        ax.text(j, value, str(value), ha='center', va='bottom', color='black', fontweight='bold', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# Group and sum the data by 'Incident year'
grouped_data = migrant_df.groupby('Incident year')[['Number of Dead', 'Minimum Estimated Number of Missing']].sum()

# Extract incident years and corresponding data
incident_years = grouped_data.index
number_of_dead = grouped_data['Number of Dead']
min_estimated_missing = grouped_data['Minimum Estimated Number of Missing']

# Create a stacked bar chart
fig, ax = plt.subplots()
bars1 = ax.bar(incident_years, number_of_dead, label='Number of Dead')
bars2 = ax.bar(incident_years, min_estimated_missing, bottom=number_of_dead, label='Minimum Estimated Number of Missing')

# Add values to the stacked bars
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black', fontsize=8)

for bar1, bar2 in zip(bars1, bars2):
    yval1 = bar1.get_height()
    yval2 = bar2.get_height()
    ax.text(bar1.get_x() + bar1.get_width()/2, yval1 + yval2, round(yval2, 2), ha='center', va='bottom', color='black', fontsize=8)

ax.set_xlabel('Incident Year', fontweight='bold', fontsize=10)
ax.set_ylabel('Number of missing and dead migrants',  fontweight='bold', fontsize=10)
ax.set_title('Migrant Incidents: Deaths and Missing Persons',  fontweight='bold', fontsize=14,  color='blue')
ax.legend()
ax.set_xticks(incident_years)
ax.set_xticklabels(incident_years, rotation=45)

plt.tight_layout()
plt.show()
```
